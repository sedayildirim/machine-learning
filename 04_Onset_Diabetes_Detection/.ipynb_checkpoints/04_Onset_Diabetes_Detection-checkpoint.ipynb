```python
import pandas as pd
import numpy as np

# import the uci pima indians diabetes dataset
filelocation = "C:/Users/Indigo/Desktop/diabetes.csv"
df = pd.read_csv(filelocation)
```

```python
# Describe the dataset
df.describe()

# some 0 values for blood pressure, BMI etc. can be seen from the min values. Indicates missing data.
```

```python
# investigate some 0 cases
df[df['Glucose'] == 0]
```

```python
# Preprocess the data, mark zero values as NaN (not a number) and drop
columns = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

for col in columns:
    df[col].replace(0, np.NaN, inplace=True)
    
df.describe()
```

```python
# Drop rows with missing values
df.dropna(inplace=True)

# summarize the number of rows and columns in df
df.describe()
```

```python
# Convert dataframe to numpy array
dataset = df.values
print(dataset.shape)
```

```python
# split into input (X) and an output (Y)
X = dataset[:,0:8]
Y = dataset[:, 8].astype(int)
```

```python
print(X.shape)
print(Y.shape)
print(X[:5])
print(Y[:5])
```

```python
# Normalize the data using sklearn StandardScaler
# As some of the values are larger than others, so the algorithm will weigh them more heavily based on their size
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X)
```

```python
print(scaler)
```

```python
# Transform and display the training data
X_standardized = scaler.transform(X)

data = pd.DataFrame(X_standardized)
data.describe()
```

```python
# import necessary sklearn and keras packages
from sklearn.model_selection import GridSearchCV, KFold
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.optimizers import Adam
```

```python
# Define a function to create a model

def create_model():
    # Define type of model
    model = Sequential()
    # Add layers, first layer is always input so below are layers 2,3,and 4
    # 2: Dense layer with 8 nodes with rectified linear unit activation
    model.add(Dense(8, input_dim = 8, kernel_initializer='normal', activation='relu'))
    # 3: Dense layer with 4 nodes with rectified linear unit activation
    model.add(Dense(4, input_dim = 8, kernel_initializer='normal', activation='relu'))
    # 4: Output layer with sigmoid activation function for binary classification
    model.add(Dense(1, activation='sigmoid'))
    
    # Compile the model
    # Define optimiser with base learning rate as 0.01 - how fast the parameters of the model are updated
    adam = Adam(lr = 0.01)
    model.compile(loss = 'binary_crossentropy', optimizer = adam, metrics = ['accuracy'])
    return model

model = create_model()
print(model.summary())
```

```python
# Do a grid search for the optimal batch size and number of epochs

# import necessary packages
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.optimizers import Adam

# Define a random seed
seed = 6
np.random.seed(seed)

# Define a function to create a model
def create_model():
    # Define type of model
    model = Sequential()
    # Add layers, first layer is always input so below are layers 2,3,and 4
    # 2: Dense layer with 8 nodes with rectified linear unit activation
    model.add(Dense(8, input_dim = 8, kernel_initializer='normal', activation='relu'))
    # 3: Dense layer with 4 nodes with rectified linear unit activation
    model.add(Dense(4, input_dim = 8, kernel_initializer='normal', activation='relu'))
    # 4: Output layer with sigmoid activation function for binary classification
    model.add(Dense(1, activation='sigmoid'))
    
    # Compile the model
    # Define optimiser with base learning rate as 0.01 - how fast the parameters of the model are updated
    adam = Adam(lr = 0.01)
    model.compile(loss = 'binary_crossentropy', optimizer = adam, metrics = ['accuracy'])
    return model

# Create the model
# Pass the model fuction and how much information is printed (verbose)
model = KerasClassifier(build_fn = create_model, verbose = 1)

# Define the grid search parameters
# Batch size: How many examples to go through before updating parameters
batch_size = [10, 20, 40]
# Epoch: How long the data is trained
epochs = [10, 50, 100]

# Make a dictionary of the grid search parameters, currently a list
param_grid = dict(batch_size=batch_size, epochs=epochs)

# Build and fit the GridSearchCV with KFold cross validation (default is 3-fold)
grid = GridSearchCV(estimator = model, param_grid = param_grid, cv = KFold(random_state=seed), verbose = 10)
grid_results = grid.fit(X_standardized, Y)

# summarize the results, 0 is the first variable for .format
# prints best score and which parameters were used for that score
print("Best: {0}, using {1}".format(grid_results.best_score_, grid_results.best_params_))
# get mean of test scores
means = grid_results.cv_results_['mean_test_score']
# get standard deviation of test scores
stds = grid_results.cv_results_['std_test_score']
# get parameters of test scores
params = grid_results.cv_results_['params']
# print all
for mean, stdev, param in zip(means, stds, params):
    print('{0} ({1}) with: {2}'.format(mean, stdev, param))
```

```python
# Seems like there is some overfitting as better scores on training set than testing set so need to optimise lr and dropout
# Do a grid search for learning rate and dropout rate

# import necessary packages
from keras.layers import Dropout

# Define a random seed
seed = 6
np.random.seed(seed)

# Define a function to create a model
def create_model(learn_rate, dropout_rate):
    # Define type of model
    model = Sequential()
    # Add layers, first layer is always input so below are layers 2,3,and 4
    # 2: Dense layer with 8 nodes with rectified linear unit activation
    model.add(Dense(8, input_dim = 8, kernel_initializer='normal', activation='relu'))
    # Dropout layer - drops some neurons to let others pick up slack so not one neurone is too important
    model.add(Dropout(dropout_rate))
    # 3: Dense layer with 4 nodes with rectified linear unit activation
    model.add(Dense(4, input_dim = 8, kernel_initializer='normal', activation='relu'))
    # Dropout layer - drops some neurons to let others pick up slack so not one neurone is too important
    model.add(Dropout(dropout_rate))
    # 4: Output layer with sigmoid activation function for binary classification
    model.add(Dense(1, activation='sigmoid'))
    
    # Compile the model
    # Define optimiser with base learning rate as given - how fast the parameters of the model are updated
    adam = Adam(lr = learn_rate)
    model.compile(loss = 'binary_crossentropy', optimizer = adam, metrics = ['accuracy'])
    return model

# Create the model
# Pass the model fuction and how much information is printed (verbose)
# Use best batch size and epoch from above grid search
model = KerasClassifier(build_fn = create_model, epochs = 10, batch_size = 20, verbose = 0)

# Define the grid search parameters
# Go below and above the ones used previously
learn_rate = [0.001, 0.01, 0.1]
dropout_rate = [0.0, 0.1, 0.2]

# Make a dictionary of the grid search parameters
param_grid = dict(learn_rate=learn_rate, dropout_rate=dropout_rate)

# Build and fit the GridSearchCV
grid = GridSearchCV(estimator = model, param_grid = param_grid, cv = KFold(random_state=seed), verbose = 10)
grid_results = grid.fit(X_standardized, Y)

# Summarize the results
print("Best: {0}, using {1}".format(grid_results.best_score_, grid_results.best_params_))
means = grid_results.cv_results_['mean_test_score']
stds = grid_results.cv_results_['std_test_score']
params = grid_results.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print('{0} ({1}) with: {2}'.format(mean, stdev, param))
```

```python
# Do a grid search to optimize kernel initialization and activation functions

# import necessary packages
from keras.layers import Dropout

# Define a random seed
seed = 6
np.random.seed(seed)

# Define a function to create a model
def create_model(activation, init):
    # Define type of model
    model = Sequential()
    # Add layers, first layer is always input so below are layers 2,3,and 4
    # 2: Dense layer with 8 nodes with rectified linear unit activation
    model.add(Dense(8, input_dim = 8, kernel_initializer=init, activation=activation))
    # Dropout layer - drops some neurons to let others pick up slack so not one neurone is too important
    model.add(Dropout(0.1))
    # 3: Dense layer with 4 nodes with rectified linear unit activation
    model.add(Dense(4, input_dim = 8, kernel_initializer=init, activation=activation))
    # Dropout layer - drops some neurons to let others pick up slack so not one neurone is too important
    model.add(Dropout(0.1))
    # 4: Output layer with sigmoid activation function for binary classification
    model.add(Dense(1, activation='sigmoid'))
    
    # Compile the model
    # Define optimiser with base learning rate as given - how fast the parameters of the model are updated
    adam = Adam(lr = 0.1)
    model.compile(loss = 'binary_crossentropy', optimizer = adam, metrics = ['accuracy'])
    return model

# Create the model
# Pass the model fuction and how much information is printed (verbose)
# Use best batch size and epoch from above grid search
model = KerasClassifier(build_fn = create_model, epochs = 10, batch_size = 20, verbose = 0)

# Define the grid search parameters
# sotmax, rectified linear unit, hyperbolic tangent, and linear activations
activation = ['softmax', 'relu', 'tanh', 'linear']
init = ['uniform', 'normal', 'zero']

# Make a dictionary of the grid search parameters
param_grid = dict(activation = activation, init = init)

# Build and fit the GridSearchCV
grid = GridSearchCV(estimator = model, param_grid = param_grid, cv = KFold(random_state=seed), verbose = 10)
grid_results = grid.fit(X_standardized, Y)

# Summarize the results
print("Best: {0}, using {1}".format(grid_results.best_score_, grid_results.best_params_))
means = grid_results.cv_results_['mean_test_score']
stds = grid_results.cv_results_['std_test_score']
params = grid_results.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print('{0} ({1}) with: {2}'.format(mean, stdev, param))
```

```python
# Do a grid search to find the optimal number of neurons in each hidden layer
# import necessary packages
from keras.layers import Dropout

# Define a random seed
seed = 6
np.random.seed(seed)

# Define a function to create a model
def create_model(neuron1, neuron2):
    # Define type of model
    model = Sequential()
    # Add layers, first layer is always input so below are layers 2,3,and 4
    # 2: Dense layer with given nodes with rectified linear unit activation
    model.add(Dense(neuron1, input_dim = 8, kernel_initializer='uniform', activation='relu'))
    # Dropout layer - drops some neurons to let others pick up slack so not one neurone is too important
    model.add(Dropout(0.1))
    # 3: Dense layer with given nodes with rectified linear unit activation
    model.add(Dense(neuron2, input_dim = neuron1, kernel_initializer='uniform', activation='relu'))
    # Dropout layer - drops some neurons to let others pick up slack so not one neurone is too important
    model.add(Dropout(0.1))
    # 4: Output layer with sigmoid activation function for binary classification
    model.add(Dense(1, activation='sigmoid'))
    
    # Compile the model
    # Define optimiser with best learning rate
    adam = Adam(lr = 0.1)
    model.compile(loss = 'binary_crossentropy', optimizer = adam, metrics = ['accuracy'])
    return model

# Create the model
# Pass the model fuction and how much information is printed (verbose)
# Use best batch size and epoch from above grid search
model = KerasClassifier(build_fn = create_model, epochs = 10, batch_size = 20, verbose = 0)

# Define the grid search parameters
neuron1 = [4, 8, 16]
neuron2 = [2, 4, 8]

# Make a dictionary of the grid search parameters
param_grid = dict(neuron1 = neuron1, neuron2 = neuron2)

# Build and fit the GridSearchCV
# added refit = True here as last optimisation step, so at the end the model is retrained with new parameters found
grid = GridSearchCV(estimator = model, param_grid = param_grid, cv = KFold(random_state=seed), refit = True, verbose = 10)
grid_results = grid.fit(X_standardized, Y)

# Summarize the results
print("Best: {0}, using {1}".format(grid_results.best_score_, grid_results.best_params_))
means = grid_results.cv_results_['mean_test_score']
stds = grid_results.cv_results_['std_test_score']
params = grid_results.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print('{0} ({1}) with: {2}'.format(mean, stdev, param))
```

```python
# generate predictions with optimal hyperparameters
y_pred = grid.predict(X_standardized)
```

```python
print(y_pred.shape)
```

```python
print(y_pred[:5])
```

```python
# Generate a classification report
from sklearn.metrics import classification_report, accuracy_score

print(accuracy_score(Y, y_pred))
print(classification_report(Y, y_pred))

# 81.9% accuracy
# precision looks at true positives and false positives
# recall is the ability to find all positive samples
```
